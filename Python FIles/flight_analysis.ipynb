# Monthly Trends: Flights and Passengers

```python
import pandas as pd
import numpy as np
import lightningchart as lc

with open('D:/Computer Aplication/WorkPlacement/Projects/shared_variable.txt', 'r') as f:
    mylicensekey = f.read().strip()
lc.set_license(mylicensekey)

monthly_trends = data.groupby(['Year', 'Month'])[['Total_Flights', 'Total_Passengers']].sum().reset_index()

# Normalize data with logarithmic scaling
monthly_trends['Log_Total_Flights'] = np.log1p(monthly_trends['Total_Flights'])
monthly_trends['Log_Total_Passengers'] = np.log1p(monthly_trends['Total_Passengers'])

# Create LightningChart instance
chart = lc.ChartXY().set_title('Monthly Trends: Flights and Passengers (Logarithmic Normalization)')

# Configure X-Axis as Month Index
x_axis = chart.get_default_x_axis().set_title('Day Index').set_tick_strategy("Numeric")

# Configure Y-Axis
y_axis = chart.get_default_y_axis().set_title('Log Count (Log1p)').set_tick_strategy("Numeric")

# Create Line Series for Total Flights
total_flights_series = chart.add_point_line_series().set_name('Log Total Flights')
total_flights_series.add(
    x=np.arange(len(monthly_trends)),
    y=monthly_trends['Log_Total_Flights']
).set_line_color(lc.Color(0, 102, 255)).set_point_shape('circle').set_point_size(5)  # Blue

# Create Line Series for Total Passengers
total_passengers_series = chart.add_point_line_series().set_name('Log Total Passengers')
total_passengers_series.add(
    x=np.arange(len(monthly_trends)),
    y=monthly_trends['Log_Total_Passengers']
).set_line_color(lc.Color(255, 102, 0)).set_point_shape('triangle').set_point_size(5)  # Orange

chart.add_legend().add(total_flights_series).add(total_passengers_series)

chart.open(method="browser")
```

# Monthly Distribution of Total Passengers - Polar - Spider

```python
import pandas as pd
import numpy as np
import lightningchart as lc

with open('D:/Computer Aplication/WorkPlacement/Projects/shared_variable.txt', 'r') as f:
    mylicensekey = f.read().strip()
lc.set_license(mylicensekey)

dashboard = lc.Dashboard(
    theme=lc.Themes.Light,
    rows=2,
    columns=2
)

# -------------------- Chart 1: Monthly Distribution of Total Passengers --------------------

# Aggregate data by Month
monthly_passengers = data.groupby('Month')['Total_Passengers'].sum()

# Prepare data for polar chart
angles = [i * (360 / 12) for i in range(12)]  # 12 months distributed across 360 degrees
values = monthly_passengers.values.astype(float)

chart1 = dashboard.PolarChart(row_index=0, column_index=0)
chart1.set_title("Monthly Distribution of Total Passengers")

series1 = chart1.add_point_series()
series1.set_data([{'angle': angle, 'amplitude': value} for angle, value in zip(angles, values)])
series1.set_point_shape('circle').set_point_size(10)

# -------------------- Chart 2: Comparison of Scheduled vs Charter Flights --------------------

# Select airports and aggregate data
selected_airports = ['MSP', 'IAH', 'ATL', 'CLT', 'IAD', 'JFK']
filtered_data = data[data['usg_apt'].isin(selected_airports)]
aggregated_data = filtered_data.groupby('usg_apt')[['Scheduled_Flights', 'Charter_Flights']].sum()

# Apply logarithmic normalization
scheduled_flights_log = np.log1p(aggregated_data['Scheduled_Flights'].values)
charter_flights_log = np.log1p(aggregated_data['Charter_Flights'].values)
max_log_value = max(scheduled_flights_log.max(), charter_flights_log.max())
scheduled_flights_normalized = (scheduled_flights_log / max_log_value) * 100
charter_flights_normalized = (charter_flights_log / max_log_value) * 100

chart2 = dashboard.SpiderChart(row_index=0, column_index=1)
chart2.set_title("Scheduled vs Charter Flights (Log Scale)")
categories = aggregated_data.index.tolist()

# Add axes (airports)
for airport in categories:
    chart2.add_axis(airport)

# Add data series
series2_scheduled = chart2.add_series()
series2_scheduled.set_name("Scheduled Flights").add_points(
    [{'axis': categories[i], 'value': value} for i, value in enumerate(scheduled_flights_normalized)]
)

series2_charter = chart2.add_series()
series2_charter.set_name("Charter Flights").add_points(
    [{'axis': categories[i], 'value': value} for i, value in enumerate(charter_flights_normalized)]
)

chart2.set_web_mode('polygon').set_axis_interval(center=0, edge=100)
chart2.add_legend().add(series2_scheduled).add(series2_charter)

# -------------------- Chart 3: Seasonal Trends: Flights vs Passengers --------------------

# Aggregate data by month
seasonal_trends = data.groupby('Month')[['Total_Flights', 'Total_Passengers']].sum()

# Apply logarithmic normalization
log_flights = np.log1p(seasonal_trends['Total_Flights'].values)
log_passengers = np.log1p(seasonal_trends['Total_Passengers'].values)
log_max_value = max(log_flights.max(), log_passengers.max())
normalized_flights = (log_flights / log_max_value) * 100
normalized_passengers = (log_passengers / log_max_value) * 100

chart3 = dashboard.SpiderChart(row_index=1, column_index=0)
chart3.set_title("Seasonal Trends: Flights vs Passengers (Logarithmic)")
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Add axes (months)
for month in month_labels:
    chart3.add_axis(month)

# Add data series
series3_flights = chart3.add_series()
series3_flights.set_name("Total Flights").add_points(
    [{'axis': month_labels[i], 'value': value} for i, value in enumerate(normalized_flights)]
)

series3_passengers = chart3.add_series()
series3_passengers.set_name("Total Passengers").add_points(
    [{'axis': month_labels[i], 'value': value} for i, value in enumerate(normalized_passengers)]
)

chart3.set_web_mode('polygon').set_axis_interval(center=0, edge=100)
chart3.add_legend().add(series3_flights).add(series3_passengers)

# -------------------- Chart 4: Carrier Performance Comparison --------------------

# Aggregate data by carrier
carrier_metrics = data.groupby('carrier').agg({
    'Total_Passengers': 'sum',
    'Total_Flights': 'sum',
    'Charter_Flights': 'sum'
}).reset_index()

# Normalize metrics (min-max scaling)
normalized_metrics = carrier_metrics.iloc[:, 1:].apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=0)
normalized_metrics['carrier'] = carrier_metrics['carrier']
metrics = ['Total Passengers', 'Total Flights', 'Charter Flights']
carriers = normalized_metrics['carrier'].tolist()

chart4 = dashboard.SpiderChart(row_index=1, column_index=1)
chart4.set_title("Carrier Performance Comparison")

# Add axes for metrics
for metric in metrics:
    chart4.add_axis(metric)

# Add a series for each carrier
for i, carrier in enumerate(carriers):
    values = normalized_metrics.iloc[i, :-1].tolist()
    data_points = [{'axis': metrics[j], 'value': value} for j, value in enumerate(values)]
    
    series4 = chart4.add_series()
    series4.set_name(carrier).add_points(data_points)

chart4.set_axis_interval(center=0, edge=1).set_web_count(10)

dashboard.open(method="browser")
```

# 3D Line Chart

```python
import time
import trimesh
import lightningchart as lc

with open('D:/Computer Aplication/WorkPlacement/Projects/shared_variable.txt', 'r') as f:
    mylicensekey = f.read().strip()
lc.set_license(mylicensekey)

data = data[data['Total_Passengers'] > 0]
grouped_data = data.groupby(['Year', 'usg_apt']).agg({'Total_Passengers': 'sum'}).reset_index()

# Ensure all values are converted to native Python types
grouped_data['Year'] = grouped_data['Year'].astype(float)
grouped_data['Total_Passengers'] = grouped_data['Total_Passengers'].astype(int)

# Selected airports
selected_airports = ['AEX', 'ABE', 'CID', 'CHS']

# Create the 3D Chart
chart = lc.Chart3D(title="Passenger Traffic by Airport", theme=lc.Themes.Dark)

# Add 3D Line Series and airplane models for each airport
airport_series = {}
airplane_models = {}
airplane_obj_path = '../Dataset/AirplaneForFreeobj.obj'  # Replace with actual path
airplane_scene = trimesh.load(airplane_obj_path)

if isinstance(airplane_scene, trimesh.Scene):
    airplane_mesh = airplane_scene.dump(concatenate=True)
else:
    airplane_mesh = airplane_scene

airplane_vertices = airplane_mesh.vertices.flatten().tolist()
airplane_indices = airplane_mesh.faces.flatten().tolist()
airplane_normals = airplane_mesh.vertex_normals.flatten().tolist()

# Function to generate unique colors
def generate_color(index, total):
    hue = index / total  # Generate a value between 0 and 1
    red = int(255 * (1 - hue))
    green = int(255 * hue)
    blue = int(255 * (0.5 - abs(hue - 0.5)))
    return lc.Color(red, green, blue)

# Create series and airplane models with unique colors
for i, airport in enumerate(selected_airports):
    if i == 0:
        color = lc.Color(255, 255, 0)  # Yellow
    elif i == 1:
        color = lc.Color(255, 165, 0)  # Orange
    elif i == 2:
        color = lc.Color(255, 0, 0)  # Red
    else:
        color = generate_color(i, len(selected_airports))

    # Create and configure line series
    series = chart.add_line_series().set_line_color(color).set_line_thickness(3)
    series.set_name(airport)
    airport_series[airport] = series

    # Create and configure airplane models
    airplane = chart.add_mesh_model().set_color(color)
    airplane.set_model_geometry(vertices=airplane_vertices, indices=airplane_indices, normals=airplane_normals)
    airplane.set_scale(0.001).set_model_rotation(0, 90, 0)
    airplane_models[airport] = airplane

# Initialize data structures
years = sorted(grouped_data['Year'].unique())
airport_data = {}
for airport in selected_airports:
    airport_data[airport] = [{'x': year, 'y': 0, 'z': selected_airports.index(airport)} for year in years]
    airport_series[airport].add(airport_data[airport])

# Axis titles
chart.get_default_x_axis().set_title('Year')
chart.get_default_y_axis().set_title('Passengers').set_interval(0, 28000, stop_axis_after=True)
z_axis = chart.get_default_z_axis().set_title('Airports').set_tick_strategy('Empty')

for i, airport in enumerate(selected_airports):
    custom_tick = z_axis.add_custom_tick()
    custom_tick.set_value(i)
    custom_tick.set_text(airport)

# Animation Function (Using Simple Loop)
def animate_chart():
    for i, year in enumerate(years):
        print(f"Animating year: {year}")
        for airport in selected_airports:
            airport_traffic = grouped_data[
                (grouped_data['usg_apt'] == airport) & (grouped_data['Year'] == year)
            ]
            
            if not airport_traffic.empty:
                total_value = int(airport_traffic['Total_Passengers'].values[0])
            else:
                total_value = 0

            # Debugging print statements
            print(f"Airport: {airport}, Year: {year}, Passengers: {total_value}")

            # Update airport data
            airport_data[airport][i]['y'] = total_value

            # Clear and re-add data for the series
            print(f"Clearing and adding data for {airport}")
            airport_series[airport].clear()
            airport_series[airport].add(airport_data[airport][:i + 1])

            airplane_models[airport].set_model_location(year, total_value, selected_airports.index(airport))
            print(f"Airplane for {airport} moved to Year={year}, Passengers={total_value}")

        time.sleep(1)

chart.open(live=True, method="browser")

animate_chart()
```

```python
import lightningchart as lc
import pandas as pd
import random

with open('D:/Computer Aplication/WorkPlacement/Projects/shared_variable.txt', 'r') as f:
    mylicensekey = f.read().strip()
lc.set_license(mylicensekey)

data = data.dropna(subset=['Latitude', 'Longitude'])


dashboard = lc.Dashboard(columns=1, rows=1, theme=lc.Themes.Light)

# --- Map Chart ---
map_chart = dashboard.MapChart(
    column_index=0,
    row_index=0
)
map_chart.set_title("")
# --- Bubble Chart ---
bubble_chart = dashboard.ChartXY(
    column_index=0,
    row_index=0,
    title="Total Flights and Passengers by Location"
)
bubble_chart.set_padding({'left': 25, 'right':88})

x_axis = bubble_chart.get_default_x_axis()
x_axis.set_title("Longitude")
x_axis.set_interval(-180, 180, stop_axis_after=True)

y_axis = bubble_chart.get_default_y_axis()
y_axis.set_title("Latitude")
y_axis.set_interval(-80, 80, stop_axis_after=True)

# Add bubble series
series = bubble_chart.add_point_series(
    sizes=True,
    rotations=True,
    lookup_values=True,
)

# Prepare data
max_flights = data['Total_Flights'].max()
max_passengers = data['Total_Passengers'].max()
x_values = []
y_values = []
lookup_values = []
rotations = []
sizes = []

for _, row in data.iterrows():
    longitude = row['Longitude']
    latitude = row['Latitude']
    total_flights = row['Total_Flights']
    if pd.notna(longitude) and pd.notna(latitude) and total_flights > 0:
        scaled_size_total_flights = 10 + (total_flights / max_flights) * 100
        scaled_size_total_passengers = 10 + (row['Total_Passengers'] / max_passengers) * 100
        x_values.append(longitude)
        y_values.append(latitude)
        sizes.append(scaled_size_total_flights)
        rotations.append(random.randint(0, 360))
        lookup_values.append(scaled_size_total_passengers)

series.append_samples(
    x_values=x_values,
    y_values=y_values,
    sizes=sizes,
    lookup_values=lookup_values
)

# Apply palette coloring
series.set_individual_point_color_enabled(True)
series.set_palette_point_coloring(
    steps=[
        {'value': 0.0, 'color': lc.Color('yellow')},
        {'value': max(lookup_values) / 2, 'color': lc.Color('orange')},
        {'value': max(lookup_values), 'color': lc.Color('red')},
    ],
    look_up_property='value',
    percentage_values=False
)

dashboard.open(method="browser")
```

# Bar chart - Line chart

```python
import pandas as pd
import lightningchart as lc
import numpy as np

with open('D:/Computer Aplication/WorkPlacement/Projects/shared_variable.txt', 'r') as f:
    mylicensekey = f.read().strip()
lc.set_license(mylicensekey)

data_bar = pd.read_csv('air traffic.csv')

# Combine 'Year' and 'Month' to create a datetime column
data_bar['Date'] = pd.to_datetime(data_bar[['Year', 'Month']].assign(DAY=1))

# Convert numerical columns from strings with commas to floats
for col in ['Dom_Pax', 'Int_Pax', 'Pax', 'Dom_Flt', 'Int_Flt', 'Flt', 
            'Dom_RPM', 'Int_RPM', 'RPM', 'Dom_ASM', 'Int_ASM', 'ASM']:
    data_bar[col] = data_bar[col].str.replace(',', '').astype(float)

# Create a dashboard
dashboard = lc.Dashboard(
    columns=1,
    rows=1,
    theme=lc.Themes.Light
)

# Create a bar chart
bar_chart = dashboard.BarChart(
    column_index=0,
    row_index=0
)

bar_chart.set_title('Monthly Airline Traffic: Flights and Load Factor')
bar_chart_data = [
    {'category': row['Date'].strftime('%Y-%m'), 'value': row['Flt']}
    for _, row in data_bar.iterrows()
]

bar_chart.set_data(bar_chart_data).set_sorting('disabled').set_value_label_display_mode('hidden')

bar_chart.set_palette_colors(
    steps=[
        {'value': 0.0, 'color': lc.Color(255, 0, 0, 128)},
        {'value': max(data_bar['Flt']), 'color': lc.Color(0, 255, 255, 128)},   
    ]
)

# Create a line chart
line_chart = dashboard.ChartXY(
    column_index=0,
    row_index=0
)
x_axis = line_chart.get_default_x_axis().set_tick_strategy('DateTime')
y_axis = line_chart.get_default_y_axis().set_interval(0, 100, stop_axis_after=True)

line_series_LF = line_chart.add_point_line_series().set_name('Load Factor (%)')
line_series_LF.set_point_shape(('circle')).set_point_color(lc.Color(204, 0, 0))
line_chart.set_title('Monthly Airline Traffic: Flights and Load Factor')
line_chart.get_default_x_axis().set_title('Date')
line_chart.get_default_y_axis().set_title('Load Factor (%)')

line_series_LF.append_samples(
    x_values=data_bar['Date'].astype(np.int64) // 10**6,
    y_values=data_bar['LF']
)
line_series_LF.set_draw_order(1).set_line_color(lc.Color(204, 0, 0))

y_axis_RPM = line_chart.add_y_axis(opposite=True)
y_axis_RPM.set_title('Engine RPM (RPM)').set_interval(1000000, 105000000, stop_axis_after=True)
line_series_RPM = line_chart.add_point_line_series(x_axis=x_axis, y_axis=y_axis_RPM).set_name('Engine RPM (RPM)')
line_series_RPM.set_point_shape(('diamond')).set_point_color(lc.Color('blue'))

line_series_RPM.append_samples(
    x_values=data_bar['Date'].astype(np.int64) // 10**6,
    y_values=data_bar['RPM']
)
line_series_RPM.set_draw_order(2).set_line_color(lc.Color('blue'))

legend = line_chart.add_legend()
legend.add(line_series_LF).add(line_series_RPM)
legend.set_margin(140).set_dragging_mode('draggable')

dashboard.open(method='browser')
```

```python
import lightningchart as lc
import numpy as np
import pandas as pd

with open('D:/Computer Aplication/WorkPlacement/Projects/shared_variable.txt', 'r') as f:
    mylicensekey = f.read().strip()
lc.set_license(mylicensekey)

data = data.dropna(subset=['Latitude', 'Longitude'])

us_lat_lng = data.groupby('usg_apt')[['Latitude', 'Longitude']].mean().to_dict('index')
fg_lat_lng = data.groupby('fg_apt')[['Latitude', 'Longitude']].mean().to_dict('index')

top_flights_country = (
    data.groupby(['Country_ISO'])
    .agg({'Total_Flights': 'sum'})
    .sort_values(by='Total_Flights', ascending=False)
    .reset_index()
)

region_data = [
    {'ISO_A3': row['Country_ISO'], 'value': float(row['Total_Flights'])}
    for _, row in top_flights_country.iterrows()
]

top_routes = data.groupby(['usg_apt', 'fg_apt']).agg(
    {'Total_Passengers': 'sum', 'Total_Flights': 'sum'}
).sort_values(by='Total_Passengers', ascending=False).head(100).reset_index()

top_routes['Total_Passengers'] = top_routes['Total_Passengers'].astype(float)
top_routes['Total_Flights'] = top_routes['Total_Flights'].astype(float)

min_flights = top_routes['Total_Flights'].min()
max_flights = top_routes['Total_Flights'].max()

dashboard = lc.Dashboard(columns=1, rows=1, theme=lc.Themes.Light)

map_chart = dashboard.MapChart(column_index=0, row_index=0).set_title("")

map_chart.invalidate_region_values(region_data)
map_chart.set_palette_coloring(
    steps=[
        {'value': min_flights, 'color': lc.Color(153, 204, 255)},
        {'value': (min_flights + max_flights) / 2, 'color': lc.Color(153, 255, 51)},
        {'value': max_flights, 'color': lc.Color(255, 0, 0)} 
    ],
    look_up_property='value',
    interpolate=True
)

line_chart = dashboard.ChartXY(column_index=0, row_index=0).set_padding({'left': 48, 'right': 85}).set_title("Popular Airline Routes")

x_axis = line_chart.get_default_x_axis().set_interval(-180, 180, stop_axis_after=True)
y_axis = line_chart.get_default_y_axis().set_interval(-84, 78, stop_axis_after=True)

route_series = line_chart.add_line_series(lookup_values=True).set_dashed("Dashed").set_line_thickness(1)

route_series.set_palette_line_coloring(
    steps=[
        {'value': min_flights, 'color': lc.Color(0, 30, 179)}, 
        {'value': (min_flights + max_flights) / 2, 'color': lc.Color(255, 255, 51)},
        {'value': max_flights, 'color': lc.Color(153, 0, 0)} 
    ],
    look_up_property='value',
    interpolate=True
)

for _, row in top_routes.iterrows():
    us_airport = row['usg_apt']
    fg_airport = row['fg_apt']
    total_flights = row['Total_Flights']

    if us_airport in us_lat_lng and fg_airport in fg_lat_lng:
        origin = us_lat_lng[us_airport]
        destination = fg_lat_lng[fg_airport]

        route_series.append_samples(
            x_values=[origin['Longitude'], destination['Longitude']],
            y_values=[origin['Latitude'], destination['Latitude']],
            lookup_values=[total_flights, total_flights]
        )

dashboard.open(method="browser")
```
